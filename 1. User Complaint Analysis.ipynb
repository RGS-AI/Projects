```python
df['Created Date']=pd.to_datetime(df['Created Date'])    #to make the data in a right format for the created date
```

```python
plt.figure(figsize=(15,12))        #plotting city wise complaint with the size of the plt being 15 & 12
sns.countplot(data=df,y='City')
plt.title('City wise total complaint')
plt.show()
```
